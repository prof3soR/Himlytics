```python
from newsapi import NewsApiClient

# Replace 'YOUR_API_KEY' with your actual News API key
newsapi = NewsApiClient(api_key='3e89568a8389414494ac2229743e1f41')
```

```python
# Define the query parameters
query = 'Himachal Pradesh'
from_date = '2025-04-01'  # Start date (YYYY-MM-DD)
to_date = '2025-04-12'    # End date (YYYY-MM-DD)
language = 'en'           # Language of articles
sort_by = 'publishedAt'   # Order by publication date

# Fetch articles
all_articles = newsapi.get_everything(q=query,
                                      from_param=from_date,
                                      to=to_date,
                                      language=language,
                                      sort_by=sort_by)

# Extract and print the relevant information
for article in all_articles['articles']:
    print(f"Title: {article['title']}")
    print(f"Description: {article['description']}")
    print(f"Published At: {article['publishedAt']}")
    print(f"URL: {article['url']}\n")
```
